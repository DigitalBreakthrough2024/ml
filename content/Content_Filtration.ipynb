#### Изменение даты загрузки в количество дней после загрузки

```python
from datetime import datetime


# Функция для перевода даты публикации в число дней с момента публикации
def days_since_publication(row):
    pub_datetime = pd.to_datetime(row['v_pub_datetime'])
    current_date = datetime.now(pub_datetime.tzinfo) # Make current_date timezone-aware
    row['days_since_publication'] = (current_date - pub_datetime).days
    return row
```

```python
df_stat = df_stat.apply(days_since_publication, axis=1)

df_stat.head()
```

#### Векторизация текстовых признаков

```python
from pymystem3 import Mystem

mystem = Mystem()

def lemmatize_text(text):
    lemmatized_words = mystem.lemmatize(str(text)) # Cast text to string type
    lemmatized_text = ' '.join([word for word in lemmatized_words if word.isalpha()])
    return lemmatized_text

for column in text_columns_no_author:
    df_stat[column] = df_stat[column].apply(lemmatize_text) # Apply the function to each column individually

df_stat[text_columns_no_author].head()
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
nltk.download('stopwords')
nltk.download('punkt')

stop_words = set(stopwords.words('russian'))

def delete_stop_words(text):
  filtered_sentence = []
  word_tokens = word_tokenize(text)
  for w in word_tokens:
    if w not in stop_words:
        filtered_sentence.append(w)
  return ' '.join(filtered_sentence)

for column in text_columns_no_author:
    df_stat[column] = df_stat[column].apply(delete_stop_words) # Apply the function to each column individually

df_stat[text_columns_no_author].head()
```

```python
csv_file_path = '/content/drive/MyDrive/ColabFiles/Datasets/train_dataset_cold_start_train/video_stat_50k_emb.csv'
df_stat.to_csv(csv_file_path, index=False)
```

```python
from transformers import AutoTokenizer, AutoModel
import torch

# Загружаем токенизатор и модель RuBERT tiny
tokenizer = AutoTokenizer.from_pretrained('cointegrated/rubert-tiny2')
model = AutoModel.from_pretrained('cointegrated/rubert-tiny2')

# Функция для создания эмбеддингов
def get_embedding(text):
    inputs = tokenizer(text, return_tensors='pt', padding=True, truncation=True, max_length=512)
    with torch.no_grad():
        outputs = model(**inputs)
    return outputs.last_hidden_state.mean(dim=1).squeeze()

# Пример обработки строки DataFrame
def process_row(row):
    category = row['category_id']
    title_description = f"{row['title']} {row['description']}"

    category_emb = get_embedding(category)
    title_desc_emb = get_embedding(title_description)

    # Объединение эмбеддингов
    combined_embedding = torch.cat((category_emb, title_desc_emb), dim=0)

    return category_emb, title_desc_emb, combined_embedding

# Применяем к DataFrame
df_stat['categoty_embedding'] = df_stat.apply(lambda row: process_row(row)[0], axis=1).tolist()
df_stat['description_embedding'] = df_stat.apply(lambda row: process_row(row)[1], axis=1).tolist()
df_stat['combined_embedding'] = df_stat.apply(lambda row: process_row(row)[2], axis=1).tolist()
```
